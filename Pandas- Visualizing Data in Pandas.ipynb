# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\10-Me22\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
#plt.style.use('classic')
```

```python
df.plot()
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Ratings')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', subplots = True)
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar',stacked = True)
```

```python
df.plot.barh()
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 600, c = 'Green')
```

```python
#histogram 
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize=(7,7))
```
